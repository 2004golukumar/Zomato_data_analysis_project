```python



df.drop_duplicates(inplace = True)
df.shape
```

```python
#cleaning Rate Column
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
#Cleaning Rest Type Column
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#Making Rest Types less than 1000 in frequency as others
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['dish_liked'] = df['dish_liked'].str.strip()
```

```python
df=df.replace('np.NaN','Not_available')
```

```python
#Cleaning Cuisines Column
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
#approx_costfor_two_people rename to  Cost2Plates
df.rename(columns = {'approx_costfor_two_people':'Cost2plates'}, inplace = True)
df.head()
```

```python
#Removing , from Cost2Plates Column
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
# Binary Encoding
binary_map = {'Yes': 1, 'No': 0}
df['online_order'] = df['online_order'].map(binary_map)
df['book_table'] = df['book_table'].map(binary_map)
```

```python
df['rate'] = df['rate'].astype(int)
df['votes'] = df['votes'].astype(float)
```

```python
# Italian Restaurants Map:
# Filter Italian restaurants
italian_df = merged_df[merged_df['cuisines'].str.contains('Italian', na=False, case=False)]

# Drop missing coordinates
italian_df = italian_df.dropna(subset=['latitude', 'longitude'])

# Create map
italian_map = folium.Map(location=[12.9716, 77.5946], zoom_start=12)

# Add markers
for _, row in italian_df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row.get('name', 'Restaurant'),
        icon=folium.Icon(color='green', icon='cutlery', prefix='fa')
    ).add_to(italian_map)

# Save map
italian_map.save("italian_restaurants_map.html")
```

```python
# Step 1: Filter restaurants with more than 1000 votes
top_rated_df = merged_df[merged_df['votes'] > 1000]

# Step 2: Group by 'rest_type' and find the average rating
rest_type_rating = top_rated_df.groupby('rest_type')['rate'].mean()

# Step 3: Sort the restaurant types by rating in descending order
sorted_rest_types = rest_type_rating.sort_values(ascending=False)
top_rest_type = sorted_rest_types.head(1)
print(top_rest_type)
```
